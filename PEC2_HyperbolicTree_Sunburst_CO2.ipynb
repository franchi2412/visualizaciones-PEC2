```python
import pandas as pd
import plotly.express as px
from pathlib import Path

# 1) Cargar datos
df = pd.read_csv("co2_by_continent_country_2023_cleaned.csv")

# 2) Crear el sunburst
fig = px.sunburst(
    df,
    path=["Continent", "Country"],
    values="CO2_Emissions",
    color="Continent",
    title="Emisiones de CO₂ por continente y país (2023)"
)
fig.update_traces(textinfo="label+percent entry")

# 3) Guardar a HTML (listo para publicar)
out = Path("sunburst_co2_2023.html")
fig.write_html(out, include_plotlyjs="cdn", full_html=True)
```

```python
from pyvis.network import Network
import pandas as pd

df = pd.read_csv("co2_by_continent_country_2023_cleaned.csv")

net = Network(height="700px", width="100%", directed=True, bgcolor="#ffffff")

net.add_node("World", size=20, color="#222222")

for cont in df["Continent"].unique():
    net.add_node(cont, color="#ffa500")
    net.add_edge("World", cont)
    for _, row in df[df["Continent"] == cont].iterrows():
        net.add_node(row["Country"], title=f"{row['CO2_Emissions']:.0f} MtCO₂")
        net.add_edge(cont, row["Country"])

net.write_html("hyperbolic_tree_pyvis.html", open_browser=False)
```
